```python
import pandas as pd
frame = pd.concat([tw2330,tw2303,tw2454,tw2317],axis=1)
frame
```

```python
frame1 = frame.pct_change()
frame2=frame1.fillna(0)
frame2
```

```python
#相關係數
frame2.corr()
```

```python
def fun1(s):
    max=s.max()*100
    min=s.min()*100
    return pd.Series([f'{max}%',f'{min}%'],index=['最高','最低'])
frame2.apply(fun1)
```
