```python
#使用codespace 要安裝中文字型
#install 要加-y
#命令列:sudo apt-get update -y
#命令列:sudo apt-get install -y fonts-wqy-zenhei
#說明網址:https://askubuntu.com/questions/490829/how-can-i-install-chinese-fonts-on-kubuntu-14-04
#字型位置:/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
#mac font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#mac font1=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=12)
#window font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
font=FontProperties(fname=r'/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)
font1=FontProperties(fname=r'/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=12)

student1={'國文':78,'英文':87,'數學':75,'歷史':64,'公民':90}
student2={'國文':75,'英文':67,'數學':66,'歷史':52,'公民':91}
student3={'國文':85,'英文':77,'數學':89,'歷史':50,'公民':45}
student4={'國文':78,'英文':98,'數學':77,'歷史':56,'公民':91}
student5={'國文':75,'英文':74,'數學':45,'歷史':78,'公民':70}
student1.keys()
student1.values()

figure=plt.figure(figsize=(8,5),dpi=300,facecolor='#dddddd')
axes=figure.add_subplot(1,1,1)
st1_score=list(student1.values())
st2_score=list(student2.values())
st3_score=list(student3.values())
axes.plot(student1.keys(),st1_score,'b*--',label='Yejun')
#axes.plot(student2.keys(),st2_score,'c*--',label='Bamby')
axes.plot(student3.keys(),student3.values(),'k*--',label='Hamin')
#axes.plot(student4.keys(),student4.values(),'r*--',label='Eunho')
#axes.plot(student5.keys(),student5.values(),'m*--',label='Noah')
axes.set_xticklabels(student1.keys(),fontproperties=font)
axes.legend()
axes.set_title("Plave的成績單",fontproperties=font)
axes.set_xlabel("科目",fontproperties=font)
axes.set_ylabel("分數",fontproperties=font)
axes.grid(axis='y')
#axes.annotate("hh",[0,90]) 註解
xlocation=axes.get_xticks()
ylocation=axes.get_yticks()
x_count=len(xlocation)
for i in range(x_count):
    x=xlocation[i]
    y1=st1_score[i]
    #y2=st2_score[i]
    y3=st3_score[i]
    axes.annotate(str(y1),[x,y1])
    #axes.annotate(str(y2),[x,y2])
    axes.annotate(str(y3),[x,y3])
plt.show()
```
