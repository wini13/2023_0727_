```python
import pandas as pd
frame=pd.read_csv('world.csv')
frame.info()
```

```python
world_frame = frame[['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數']]
world_frame
```

```python
world_frame["洲名"]
```

```python
#subscript的搜尋
asia_frame = world_frame[world_frame["洲名"] == "亞洲"]
asia_frame
```

```python
#subscript的搜尋
condition1 = (world_frame["國家"] == "台灣") & (world_frame["新增確診數"] == 0)
condition1
```

```python
world_frame[condition1]
```

```python
#用query搜尋
world_frame.query('國家=="台灣" and 新增確診數==0')
```

```python
world_frame.query('日期=="2021-08-31" and 新增確診數>50')
```

```python
world_frame.query('日期>="2020-08-31" and 國家=="台灣" and 日期<="2020-09-32"')
```

```python
taiwan=world_frame.query('國家=="台灣"')
taiwan_max=taiwan["新增確診數"].max()
taiwan_max
```

```python
taiwan.query('新增確診數==@taiwan_max')
```

```python
japan=world_frame.query('國家=="日本"')
japan_max=japan["新增確診數"].max()
japan.query("新增確診數==@japan_max")
```
