```python
import pandas as pd
data={"字串":["台北","台中"],
      "整數":[1000,3000],
      "浮點數":[3.527,2.15]}
df=pd.DataFrame(data)
display(df)
```

```python
#視覺美化
df.style\
.format(precision=2,thousands=",",decimal=".")\
.relabel_index(["列1","列2"])
```

```python
import numpy as np
data=np.random.rand(10,2)*5
df=pd.DataFrame(data,columns=['台北','台東'],index=pd.date_range(start='2023-09-30',periods=10))
df
```

```python
def feel_condition(v):
    if v<1.75:
        return"溫暖"
    elif v<2.75:
        return"潮濕"
    else:
        return"日照"
df.style\
.format(feel_condition)\
.set_caption("體感溫度")\
.background_gradient(cmap='RdPu',vmin=1,vmax=9,axis=None)
```
