```python
import pandas as pd
df=pd.read_csv('學生成績.csv')
df.index=df['學號']
df1=df.drop(columns='學號')
```

```python
df1['排名']=df1['平均'].rank(method='min',ascending=False)
df2=df1.sort_values(by='平均',ascending=False)
df2
```
