```python
import pandas as pd
df = pd.read_csv('學生成績.csv')

def less60(v,prop=''):
    return prop if v<60 else None

df.style\
.map(less60,prop='color:red',subset=["國文","英文","數學","生物","歷史"])\
.format(precision=2,subset=["平均"])
```
