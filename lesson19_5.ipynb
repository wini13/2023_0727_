```python
import numpy as np
import pandas as pd

frame = pd.read_excel('學生成績單.xlsx')
frame
```

```python
def grade(v):
    if v >= 80:
        return '優'
    elif v >= 70:
        return '甲'
    elif v >= 60:
        return '乙'
    else:
        return '丙'


frame['等級'] = frame['平均'].map(grade)
```
